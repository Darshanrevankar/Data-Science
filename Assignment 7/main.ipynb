```python
from PIL import Image, ImageDraw, ImageFont
import matplotlib.pyplot as plt

image_path = 'images\meme_5.jpg'  
img = Image.open(image_path)

draw = ImageDraw.Draw(img)

font_size = 60 
try:
    font = ImageFont.truetype("arial.ttf", font_size) 
except IOError:
    font = ImageFont.load_default()

text_1 = """ FRONT-END 
DEVELOPER..."""

text_2 = """DATA SCIENCE 
WITH GENAI """

text_position = (650, 250)
text_position_2 = (650, 800)
draw.text(text_position, text=text_1, fill=(255, 0, 0), font=font)
draw.text(text_position_2, text=text_2, fill=(255, 0, 0), font=font)
plt.imshow(img)
plt.show()
```
